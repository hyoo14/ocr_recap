```python
#tesseract안에 있는 testrain 실행을 위해 권한 줌(먼저 tesseract-4.1을 tesseract로 변경)
!chmod 755 -R /content/tesseract/src/training/tesstrain.sh
```

```python
#데이터 합쳐줌
!combine_tessdata -e /content/tessdata_best/eng.traineddata eng.lstm
```

```python
#앞서 합친 거 복사해줌
!cp /content/tessdata_best/eng.traineddata /content/tesseract/tessdata
```

```python
#tesstrain 실행
! /content/tesseract/src/training/tesstrain.sh --fonts_dir fonts \
	--fontlist '{font_name}' \
	--lang eng \
	--linedata_only \
	--langdata_dir /content/langdata_lstm \
	--tessdata_dir /content/tesseract/tessdata \
	--save_box_tiff \
	--maxpages 10 \
	--output_dir train
```

```python
#파인튜닝
%%shell

OMP_THREAD_LIMIT=16 lstmtraining \
	--continue_from eng.lstm \
	--model_output output/font_name \
	--traineddata /content/tesseract/tessdata/eng.traineddata \
	--train_listfile train/eng.training_files.txt \
	--max_iterations 1000
```

```python
#파인튜닝2
%%shell

OMP_THREAD_LIMIT=16 lstmtraining \
	--continue_from eng.lstm \
	--model_output output/font_name \
	--traineddata /content/tesseract/tessdata/eng.traineddata \
	--train_listfile train/eng.training_files.txt \
	--max_iterations 2500
```

```python
#파인튜닝3
%%shell

OMP_THREAD_LIMIT=16 lstmtraining \
	--continue_from eng.lstm \
	--model_output output/font_name \
	--traineddata /content/tesseract/tessdata/eng.traineddata \
	--train_listfile train/eng.training_files.txt \
	--max_iterations 3000
```

```python
# 트레인모델(체크포인트) 출력 디렉토리로 내보냄
%%shell

lstmtraining --stop_training \
	--continue_from output/font_name_checkpoint \
	--traineddata /content/tesseract/tessdata/eng.traineddata \
	--model_output output/font_name.traineddata
```
